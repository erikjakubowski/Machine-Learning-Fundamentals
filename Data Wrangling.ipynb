This chapter is primarly using pandas to hold data in Series and Dataframes and manipulate the structure and expose functions of the libraries that would be used to leverage or show useful metrics/reformating techniques. 

```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
cancer_data = datasets.load_breast_cancer()

df = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
df.head(5)
```

```python
#shape of data
df.shape

#info about df
df.describe()

#pin point rows
df.iloc[299:310]

#change index
df.set_index(df['mean radius'])

#select row with conditional
df[df['mean concavity'] <= 1].head(3)

#multi conditional
df[(df['mean compactness'] > .1 ) & (df['compactness error'] < .03)]

#rename column 
df.rename(columns={'radius error': 'radius err'}).head()

#get unique total
df['mean texture'].nunique()

#unique values with times it appears
df['mean texture'].value_counts()

#delete column
df.drop(['radius error'], axis=1).head(2)

#drop duplicates
df.drop_duplicates(subset=['mean radius'])

#group rows by and calc mean of each group
df.groupby('mean smoothness').mean()
```

```python
# loop over df first five
for x in df['mean concave points'][:5]:
    print(x * 100)
    
#or for different 5 as comprehension
[ x*100 for x in df['mean concave points'][10:16]]





#loop with a function
def times_ten(x):
    return x * 10

#apply function to 5
df['mean radius'].apply(times_ten)[23:28]



#apply function to groupby
df.groupby('mean symmetry').apply(lambda x : x.sum())
```
